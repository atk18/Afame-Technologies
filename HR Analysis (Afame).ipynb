```python

plt.figure(figsize=(10, 6))
plt.bar(data['Gender'], data['Department'], color='skyblue')
plt.xlabel('Gender')
plt.ylabel('Department')
plt.title('Gender by Department')

plt.xticks(rotation=45, ha='right')  
plt.tight_layout()

plt.show()
```

```python
sns.countplot(x="Gender",data=data)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(data['JobRole'], data['MonthlyIncome'], color='green')
plt.xlabel('JobRole')
plt.ylabel('MonthlyIncome')
plt.title('Monthly Income by Job Role')

plt.xticks(rotation=45, ha='right')  
plt.tight_layout()

plt.show()
```

```python

plt.figure(figsize=(10, 6))
plt.bar(data['WorkLifeBalance'], data['JobRole'], color='orange')
plt.xlabel('WorkLifeBalance')
plt.ylabel('JobRole')
plt.title('WorkLifeBalance by JobRole')

plt.xticks(rotation=45, ha='right')  
plt.tight_layout()

plt.show()
```
